```python
#jump search
'''
1.size n/sorted 
2.create a block key(which will be searhing element) 
3.search operation will be performed
arr[m]<var#(key)<arr(k+1)[m]
4.compare each jump linearly
'''
#program
import math
# time comlexcity:  0(sqrt(n))
#space complexcity: 
def jump_search(arr,target):
    if not arr:
        return -1
    n=len(arr)
    step=int(math.sqrt(n))
    prev=0
    while prev <n and arr[prev]<target:
        prev +=step
    for i in range(max(0,prev-step),min(n,prev+1)):
        if arr[i]==target:
            return i
    return -1    
arr=[1,2,3,4,5,6,7,8,9]
target=8
result=jump_search(arr,target)
print(f"Element {target} found at index: {result}")

```

```python
#exponential search
'''
1.sorrted numbers searching
2.check the arraay!=0
3.check the first element
4.find thhe range using exo while for boundry identify within boundry
5. range perform binary search
6.retur result'''

#program

def bsearch_range(arr,target,left,right):
    while left<=right:
        mid=(left+right)//2
        if arr[mid]==target:
            return mid
        elif arr[mid]<target:
            left=mid+1
        else:
            right=mid-1
    return -1

def expo_search(arr,target):
    if not arr:
        return -1
    if arr[0]==target:
        return 0
    n=len(arr)
    i=1
    while i<n and arr[i]<=target:
        i*=2
    return bsearch_range(arr,target,i//2,min(i,n-1))
arr=[2,3,4,6,7,8,9,10]
target=7
result=expo_search(arr,target)
print(f"Element {target} found at index: {result}")
```

```python
#fibonacci search
'''
1. find smallest fibo number
2.
'''
def fibsearch(arr,target):
    if not arr:
        return -1
    n=len(arr)   #initialise fib num
    fib2=0
    fib1=1
    fib=fib1+fib2
    while fib<n:  #smallest fib>=n
        fib2=fib1
        fib1=fib
        fib=fib1+fib2
    offset=-1   #initialise offset
    while fib>1:  #searching using f(0)......f(n)
        i=min(offset+fib2,n-1)
        #index value of fibnaci equavalent
        if arr[i]==target:
            return i
        elif arr[i]<target:
            offset=i
            fib=fib1
            fib1=fib2
            fib2=fib-fib1
        else:
            fib=fib
            fib1=fib1-fib2
            fib2=fib-fib1
    if fib==1 and offset+1<n and arr[offset+1]==target:
        return offset+1
    return -1
arr=[2,4,6,8,3,4,6,7]
target=8
result=fibsearch(arr,target)
print(f"Element {target} found at index: {result}")
    
        
```
